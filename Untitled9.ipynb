```julia
u(c)=c^(1-γ)/((1-γ))

ph=0.5
pl=0.5
γ=2
yl=0
yh=1

β=0.95
r=1/β
ns=100
x=range(0.2,10,length=ns)
```

```julia
using Optim
function T(v, grid, β, r, u; compute_policy = false)
    v_func = CubicSplineInterpolation(grid, v,extrapolation_bc = Line())
    Tv=similar(grid)
    σ=similar(grid)
    for i in eachindex(grid)
        X=grid[i]
        objectives = (c -> u(c)+ β*ph*(v_func.(r*(X-c)+yh ))+ β*pl*(v_func.(r*(X-c)+yl )) )
        results = maximize(objectives, 0, X) # solver result for each grid point

        Tv[i] = Optim.maximum(results)
        σ[i] = Optim.maximizer(results)
       
    end
    
     if compute_policy
        return (Tv,σ)
    else
        return (Tv)
    end
    
     
end
```

```julia

v0(k)=k;
v1=v0.(x);

ϵ=1e-5;
ρ=2*ϵ;
v=similar(x)
iter=0
valuee=plot(title="value function")
while ρ>ϵ && iter<200
    v=T(v1, x, β, r, u; compute_policy = false)
    ρ=maximum(abs.(v-v1))
    v1=v
    iter+=1
    plot!(valuee,x,v,title="value function")
    
end

v,p=T(v1, x, β, r, u; compute_policy = true);
plot(valuee)
```
